```python
import random

def scoring(user,comp):
    global comp_score,user_score
    if(user == comp):
        print('\n-------------------------------------------------\n')
        print('Its a Tie')
        print("No one get the point")
        print('\n-------------------------------------------------\n')
        
    elif(user == 'rock' and comp == 'scissor'):
        print('\n-------------------------------------------------\n')
        print("Player Won")
        print('\n-------------------------------------------------\n')
        user_score += 1

        
    elif(user == 'paper' and comp == 'rock'):
        print('\n-------------------------------------------------\n')
        print("Player Won")
        print('\n-------------------------------------------------\n')
        user_score += 1


    elif(user == 'scissor' and comp == 'paper'):
        print('\n-------------------------------------------------\n')
        print("Player Won")
        print('\n-------------------------------------------------\n')
        user_score += 1


    else:
        print('\n-------------------------------------------------\n')
        print("Computer Won")
        print('\n-------------------------------------------------\n')
        comp_score += 1
        
    print('\nScore->')
    print('Player     Computer')
    print('  ',user_score,'        ',comp_score,'\n')


op = ['Rock','Paper','Scissor']
check = ['rock','paper','scissor']

user_score = 0
comp_score = 0

ans = 'y'
while ans == 'y':
    print('\n-------------------------------------------------\n')
    print('Score->')
    print('Player     Computer')
    print('  ',user_score,'        ',comp_score,'\n')
    user = str(input("Rock Paper Scissor: "))
    comp = random.choice(op)

    if(user.lower() in check):
        print('\nYour choice->',(user.lower()).capitalize())
        print('Computers Choice->',comp,"\n")
        user = user.lower()
        comp = comp.lower()
        scoring(user,comp)

    else:
        print('Wrong Input')
        print("Retry")
    ans = input("Do you want to continue(y/n)? ")

print('\n-------------------------------------------------\n')

if(user_score>comp_score):
    print("Player Won")
    print('\n-------------------------------------------------\n')
    
elif(user_score<comp_score):
    print('Computer won')
    print('\n-------------------------------------------------\n')
    
else:
    print('Its a Tie')
    print('\n-------------------------------------------------\n')
    
print('Score->')
print('Player     Computer')
print('  ',user_score,'        ',comp_score,'\n')
print('\n-------------------------------------------------\n')  
```
